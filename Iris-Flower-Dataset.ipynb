```python
numeric_columns = Iris_flower_dataset.drop(columns = 'species', axis = 1)
numeric_columns
```

```python
corelation = numeric_columns.corr()
fig, axis = plt.subplots(figsize = (5,5))
sns.heatmap(corelation, annot = True, ax = axis, cmap = 'BuPu')
```

```python
label = LabelEncoder()
```

```python
numeric_columns.head()
```

```python
X = numeric_columns
```

```python
Iris_flower_dataset['species'] = label.fit_transform(Iris_flower_dataset['species'])
```

```python
Iris_flower_dataset.head(15)
```

```python
X = Iris_flower_dataset.drop(columns = 'species', axis = 1)
Y = Iris_flower_dataset['species']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
LG = LogisticRegression()
```

```python
LG.fit(X_train, Y_train)
```

```python
KNC = KNeighborsClassifier()
```

```python
KNC.fit(X_train, Y_train)
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(X_train, Y_train)
```

```python
LG_Accuracy = LG.score(X_test, Y_test)*100
KNC_Accuracy = KNC.score(X_test, Y_test)*100
DTC_Accuracy = DTC.score(X_test, Y_test)*100
```

```python
print(f"Accuracy of Iris Flower Prediction by using Logistic Regression is : {LG_Accuracy}% ")
```

```python
print(f"Accuracy of Iris Flower Prediction by using KNeighbors Classifier is : {KNC_Accuracy}% ")
```

```python
print(f"Accuracy of Iris Flower Prediction by using Decision Tree Classifier is : {DTC_Accuracy}% ")
```
